```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib

# CSV 파일 읽기
df = pd.read_csv('학교생활만족도.csv')
df.columns = df.columns.str.strip()  # 열 이름 정리

# 첫 번째 행 두 번 삭제
df = df.drop(0, axis=0).reset_index(drop=True)
df = df.drop(0, axis=0).reset_index(drop=True)

# 열 이름 매핑
column_mapping = {
    '2024': '학교 친구들은 나를 존중하고 배려해준다(전혀 그렇지 않다)',
    '2024.1': '학교 친구들은 나를 존중하고 배려해준다(그렇지 않은 편이다)',
    '2024.2': '학교 친구들은 나를 존중하고 배려해준다(그런 편이다)',
    '2024.3': '학교 친구들은 나를 존중하고 배려해준다(매우 그렇다)',
    '2024.4': '선생님은 학생들을 존중한다(전혀 그렇지 않다)',
    '2024.5': '선생님은 학생들을 존중한다(그렇지 않은 편이다)',
    '2024.6': '선생님은 학생들을 존중한다(그런 편이다)',
    '2024.7': '선생님은 학생들을 존중한다(매우 그렇다)',
    '2024.8': '나는 수업시간이 재밌다(전혀 그렇지 않다)',
    '2024.9': '나는 수업시간이 재밌다(그렇지 않은 편이다)',
    '2024.10': '나는 수업시간이 재밌다(그런 편이다)',
    '2024.11': '나는 수업시간이 재밌다(매우 그렇다)',
    '2024.12': '나는 학교에 가는 것이 즐겁다(전혀 그렇지 않다)',
    '2024.13': '나는 학교에 가는 것이 즐겁다(그렇지 않은 편이다)',
    '2024.14': '나는 학교에 가는 것이 즐겁다(그런 편이다)',
    '2024.15': '나는 학교에 가는 것이 즐겁다(매우 그렇다)'
}

# 열 이름 변경
df = df.rename(columns=column_mapping)

# 필요있는 열만 추출
respect_columns = [
    '응답자유형별(1)', '응답자유형별(2)',
    '학교 친구들은 나를 존중하고 배려해준다(전혀 그렇지 않다)',
    '학교 친구들은 나를 존중하고 배려해준다(그렇지 않은 편이다)',
    '학교 친구들은 나를 존중하고 배려해준다(그런 편이다)',
    '학교 친구들은 나를 존중하고 배려해준다(매우 그렇다)',
]

respect_df = df[respect_columns]

# 확인
respect_df.head(22)

# 그래프 스타일 설정
sns.set(style="whitegrid")

# 응답 항목만 추출
response_columns = [
    '학교 친구들은 나를 존중하고 배려해준다(전혀 그렇지 않다)',
    '학교 친구들은 나를 존중하고 배려해준다(그렇지 않은 편이다)',
    '학교 친구들은 나를 존중하고 배려해준다(그런 편이다)',
    '학교 친구들은 나를 존중하고 배려해준다(매우 그렇다)',
]

# 숫자로 변환
response_data = respect_df[response_columns].apply(pd.to_numeric, errors='coerce')

# 합계
total_responses = response_data.sum().sum()  # 전체 총합
respect_percentage = response_data.sum() / total_responses * 100  # 비율 계산

# 그래프 스타일 설정
sns.set(style="whitegrid")

# 원그래프 (파이차트) 그리기
plt.figure(figsize=(8, 8))  # 원그래프는 조금 더 정사각형 모양이 좋습니다.

# 파이차트 그리기 (수치값을 없앰)
plt.pie(respect_percentage, labels=respect_percentage.index, startangle=140, colors=sns.color_palette("Blues_d", len(respect_percentage)))

# 그래프 제목
plt.title("학교 친구들의 존중과 배려에 대한 응답 비율", fontsize=24)

# 한글 폰트 적용
plt.rcParams['font.family'] = 'NanumGothic'  # 한글 폰트 NanumGothic으로 설정

plt.tight_layout()
plt.show()
```
