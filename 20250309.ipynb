- Matplotib : 파이썬의 기본 시각화 라이브러리

    사용법 : import matplotib as plt
    기본 구성 요소 : figure, axis, plot

- 기본적인 그래프의 종류

```python
import numpy as np # 파이썬에서 연산을 담당하는 라이브러리
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
years = np.array([2000, 2005, 2010, 2015, 2020]) # x축에 넣어줄 연도
group1 = np.array([1, 4, 9, 16, 25]) # 예제 1 - 제곱값
group2 = np.array([1.5, 5, 7.5, 11, 18]) # 예제 2 - 불규칙 증가
```

```python
# 다중 선 그래프 생성
plt.plot(years, group1, label = 'Group 1', marker = "o", linestyle = "--", color = "b")
plt.plot(years, group2, label = "Group 2", marker = "s", linestyle = "-", color = "r")
```

label : 각 선의 이름을 정의해서 범례에 표시될 텍스트를 설정

marker : 선 위에 데이터의 지점을 표시하는 기호 설정

    'o' : 원형 마커
    's' : 사각형 마커

linestyle : 선 스타일 설정

    '--' : 점선
    '-' : 실선

color : 색상 설정

    'b' : blue
    'r' : red

```python
# 막대 그래프 - 카테고리별 판매량

categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78] # 막대의 길이

plt.bar(categories, values, color = ['pink', 'aqua', 'green', 'yellow'],
        edgecolor = 'white', linewidth = 1.0, alpha = 0.8, width = 0.8)
plt.title("카테고리별 판매량")
plt.xlabel("카테고리 종류")
plt.ylabel("판매량")
plt.show()
```

color : 막대 색상을 지정 (직접 입력)

edgecolor : 막대 테두리 색깔 지정

linewidth : 테두리의 두께

width : 막대의 두께

alpha : 투명도

```python
# 서브 플롯 : 그래프 안의 또 다른 그래프

# 데이터 -> 하나의 그래프에 두 개의 y축을 가진 시각화
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [30, 25, 15, 10, 5]

# 그래프 설정
fig, ax1 = plt.subplots() # 그리프(fig)와 첫 번째 축(ax1)을 생성
# ax1은 기본적인 x축과 y축을 포함하는 첫 번째 그래프 축이다.

# 첫 번째 y축
ax1.bar(x, y1, color = 'g')
ax1.set_xlabel('X 데이터')
ax1.set_ylabel('Y1 데이터', color = 'g')

ax2 = ax1.twinx() # 두 번째 축 생성
ax2.plot(x, y2, color = 'b', marker = 'o')
ax2.set_ylabel('Y2 데이터', color = 'b')

plt.title('다중 축 그래프')
plt.show()
```

# 산점도 (Scatter Plot)

    - 데이터를 x축과 y축에 값으로 표현한 그래프
    - 나이와 소득처럼 연속된 값으로 된 두 변수와 관계를 표현할 때 사용

```python
import seaborn as sns # 관계적으로 해당 라이브러리는 sns 라는 별치을 가진다.
import pandas as pd

mpg = pd.read_csv('mpg.csv')
mpg
```

```python
# x축에는 displ(배기량), y축은 hwy(고속도로 연비)를 나타낸 산점도 생성
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy') \
.set(xlim = [3, 6]) # x축 범위를 3~6으로 제한
```

**해석 : 분포로 보아 배기량이 클수록 고속도로 연비는 낮아지는 경향이 보인다**

```python
# x축 범위 3~6, y축 범위 10~30으로 제한

sns.scatterplot(data = mpg, x = 'displ', y = 'hwy') \
.set(xlim = [3, 6], ylim = [10, 30])
```

```python
# 종류별로 표식 색상 바꾸기 - drv(구동방식) 별로 색상 다르게 표현
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy', hue = 'drv')
```

해석 : f(전륜)과 4(4륜) 차들은 배기량이 커질수록 고속도로 연비가 낮아지는 경향이 있는 반면에, r(후륜) 자동차들은 특별한 비례 관계가 보이지 않는다.

# 막대 그래프(Bar Chart)

  - 데이터의 크기를 막대의 길이로 표현한 그래프
  - 성별 소득 차이처럼 집단 간 차이를 표현할 때 많이 사용한다.

# 평균 막대 그래프

```python
# 1. 집단별 평균표 생성

df_mpg = mpg.groupby('drv') \
.agg(mean_hwy = ('hwy', 'mean')) # 고속도로 연비의 평균을 산출

df_mpg
```

# as_index = False : 변수를 인덱스로 바꾸지 않고 원래대로 유지하는 옵션

  - 앞 코드의 출력 결과를 보면 집단을 나타낸 변수 drv가 인덱스로 바뀌어서 mean_hwy보다 아래 표시됨.
  - seaborn으로 그래프를 만들면 값이 변수에 담겨 있어야함

```python
df_mpg = mpg.groupby('drv', as_index = False) \
.agg(mean_hwy = ('hwy', 'mean')) # 고속도로 연비의 평균을 산출

df_mpg
```

```python
# 그래프 생성
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

# 빈도 막대 그래프

  - 여러 집단의 빈도를 비교할 때 사용
  - 집단별 빈도표를 만드는 작업을 생략이 가능함

```python
sns.countplot(data = mpg, x = 'drv')
```

```python
# 4, f, r 순서로 막대 정렬
sns.countplot(data = mpg, x = 'drv', order = ['4', 'f', 'r'])
```

```python
# drv 값을 빈도가 높은 순서로 출력
order = mpg['drv'].value_counts().index
order

sns.countplot(data=mpg, x='drv', order=order)
plt.show()
```

# Seaborn 시각화 심화 - penguins dataset 예제

  - penguins dataset은 Seaborn에 내장된 데이터 set이다.
  - seaborn 같은 경우 내장 데이터를 지원해서 처음 접하는 사람들이 공부하기 편하게 만들어져있다.

```python
# data load
import seaborn as sns

data = sns.load_dataset("penguins")
data
```

# 변수 설명
- species : 펭귄 종
- island : 서식지
- bill_length_mm : 부리 길이
- bill_depth_mm : 부리 위아래 두께

```python
# 데이터를 보니 NaN 값이 존재하는게 보인다 -> 처리를 해줘야겠다.
data = data.dropna() # dropna() 괄호를 비워놓으면 NaN 존재하는 모든 행을 지워버린다.
data
```

# HISTPLOT : 히스토그램
  - 가장 기본적으로 사용되는 히스토그램을 출력하는 plot
  - 전체 데이터를 특정 구간별 정보를 확일할 때 사용한다. (데이터 파악 부분에서)

```python
# 몸 무게의 분포를 히스토그램으로 출력
sns.histplot(data = data, x = 'body_mass_g')
```

**히스토그램의 경우 구간의 따른 카운트를 그래프에 표현하기 때문에 y는 count로 고정이 되어 있어서 x값먼 설정하면 된다.**

히스토그램은 구간의 개수에 따라 분포를 다르게 표현할 수 있는데, 이를 조정하는 옵션은 bins이다

```python
sns.histplot(data = data, x = 'body_mass_g', bins = 15, hue = 'species')
```

hue 옵션은 입력하는 변수에 따라 색을 다르게 표현하는 기능

데이터에서 종의 개수는 3개인데 그래프 내에 표현된 색상은 5개이다.

이는 겹치는 부분이 존재하기 때문에 다른 색으로 보이게 되는 것 -> 수정해서 3개의 범례로만 보고 싶은 경우는 mutiplt 옵션을 지정

```python
sns.histplot(data = data, x = 'body_mass_g', bins = 15, hue = 'species', multiple = 'stack')
```

seaborn은 matplotlib의 기능을 확장한 것임을 확인해보자.

```python
import matplotlib.pyplot as plt
# !pip install koreanize_matplotlib
# import koreanize_matplotlib

# 그래프에 title 추가
plt.figure()
plt.title('Distribution over body_mass', fontsize = 14)
sns.histplot(data = data, x = 'body_mass_g', bins = 15 , hue = 'species', multiple = 'stack')
plt.show()
```

# Violineplot : 변수에 대한 분포 자체를 보여주는 plot

상자 그림이랑 비슷하지만, 전체 분포를 보여준다는 점에서 차이가 존재한다. 상자그림과 같이 그려 평균 근처에 데이터가 얼마나 있는 전체적으로 어떻게 퍼져있는지 확인이 가능하다.

# Lineplot : 두 변수간의 관계를 확인할 수 있는 plot, 수치형 자표들 간의 경향을 파악할 때 사용한다.

```python
# 몸무게와 펭귄의 팔 길이에 대한 관계
sns.lineplot(data = data, x = 'body_mass_g', y = 'flipper_length_mm', ci = None) # ci = None : 허용범위 지우기
```

해석 : 몸무게와 팔 길이 간의 관계를 선 그래프로 통해 확인해보면, 전반적으로 몸무게가 증가할수록 팔 길이도 증가하는 경향이 있는 것을 알 수 있다. 그렇지만 증감을 반복하면서 우샹향을 하기 때문에 무조건적인 것을 아니다.

# Pointplot : 특정 수치 데이터를 error bar와 함께 출력해주는 plot

  - 수치 데이터를 다양한 각도에서 한 번에 바라보고 싶을 때 사용
  - 데이터와 에러 바를 한 번에 찍어주기 때문에 특정 지표들만 사용하는 것이 좋다.

```python
# 성별에 따른 부리 길이를 확인
sns.pointplot(data = data, x = 'sex', y = 'bill_length_mm')
```

직선은 각 분포를 표현한 것이고 직선 중간에 있는 점은 각 변수의 평균에 해당한다. 결론은 수컷의 부리가 암컷의 부리보다 길다.

Pairplot : 주어진 데이터의 각 변수들 사이의 관계를 표현하는 plot이다.

# Heatmap : 정사각형 그림에 데이터에 대한 정도 차이를 보여주느 plot

  - 열화상카메라로 사물을 찍은 것처럼 정도의 차이를 보여준다.
  - pairplot과 비슷하게 변수들간의 관계를 시각화 할 때 많이 사용된다.
  - **특히, 변수간의 상관관계를 시각화할 때 많이 사용된다.**

# 상관계수 : 어떤 x 값이 변화함에 따라 y값의 선형적으로 변화하는지를 측정한 지표

  - 0 ~ 1 사이의 값을 가진다.
  - 양수면 정비례, 음수면 반비례 관계를 가진다.

```python
# 히트맴을 그리기 위해서 상관계수를 구한다.
data2 = data[['bill_length_mm',
              'bill_depth_mm',
              'flipper_length_mm',
              'body_mass_g']]

corr = data2.corr() # 상관계숫 생성
corr
```

```python
sns.heatmap(data = corr, square = True, cmap = 'Blues')
```

```python
# 그림의 상자안에 상관계수 값을 표시
sns.heatmap(data = corr, square = True, cmap = 'Blues', annot = True)
```

몸무게와 팔 길이 사이의 상관계수가 0.8%로 강한 양의 상관관계에 있다.
